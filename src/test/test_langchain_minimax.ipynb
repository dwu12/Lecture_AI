```python
from deepagents import create_deep_agent
from dotenv import load_dotenv
from langchain_anthropic import ChatAnthropic
load_dotenv()
```

```python
model = ChatAnthropic(
    model="MiniMax-M2.7",
    temperature=0,
    max_tokens=10000,
)

agent = create_deep_agent(model=model, 
                          system_prompt = "You are a help assistant")


result = agent.invoke(
    {"messages": [{"role": "user", "content": "What's the weather in San Francisco?"}]}
)
```

```python
result['messages'][-1].content
```

```python
prompt = '''
You are a helpful prompt generator, you will generate a system prompt for my orchestrator agent. 

The basic task is to convert an Lecture audio transcribe into a human readable format. Extract key information from the transcriptions. For example, what is the task, when is it due and what is the key points. 
You will also need to generate a lecture note and save it into a PDF format. 

You will have two subagent to help you do this. 
1. transcription_agent: transcription-agent will read through the transcripts, write it into a speaker-aware way. Right now, everything is on one line. The refined scripts will be saved at some place
2. key_note agent: key note agent will take the transcription from transcirtion agent and then generate lecture note as well as the key task-time-keypoint as the pending tasks 

You should be able to answer basic questions as well 

Now, generate the system prompt for me: 
''' 

result = model.invoke(prompt)
```
