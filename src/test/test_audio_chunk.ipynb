```python
import os
from pydub import AudioSegment

def separate_audio_chunk(audio_path, max_size_mb=20):
    """
    Split or convert an audio file based on size.
    
    Args:
        audio_path: Path to the input audio file
        max_size_mb: Maximum chunk size in MB (default 20)
    
    If audio size < max_size_mb: convert to mp3 if not already mp3
    If audio size >= max_size_mb: split into chunks of max_size_mb
    """
    if not os.path.exists(audio_path):
        raise FileNotFoundError(f"Audio file not found: {audio_path}")
    
    # Get file size in MB
    file_size_mb = os.path.getsize(audio_path) / (1024 * 1024)
    
    # Get base name and directory
    base_name = os.path.basename(audio_path)
    name, ext = os.path.splitext(base_name)
    output_dir = os.path.dirname(audio_path)
    
    # Load audio
    audio = AudioSegment.from_file(audio_path)
    
    if file_size_mb < max_size_mb:
        # Convert to mp3 if not already
        output_path = os.path.join(output_dir, f"{name}.mp3") if ext.lower() != ".mp3" else audio_path
        if ext.lower() != ".mp3":
            audio.export(output_path, format="mp3")
            print(f"Converted: {output_path}")
        else:
            print(f"Audio is already MP3 and under {max_size_mb}MB: {audio_path}")
    else:
        # Split into chunks
        # Calculate chunk duration based on file size ratio
        total_duration_ms = len(audio)
        num_chunks = int((file_size_mb / max_size_mb) + 0.99)  # Ceiling division
        chunk_duration_ms = total_duration_ms // num_chunks
        
        for i in range(num_chunks):
            start = i * chunk_duration_ms
            end = start + chunk_duration_ms if i < num_chunks - 1 else total_duration_ms
            chunk = audio[start:end]
            chunk_path = os.path.join(output_dir, f"{name}_chunk_{i + 1}.mp3")
            chunk.export(chunk_path, format="mp3")
            print(f"Exported chunk {i + 1}/{num_chunks}: {chunk_path}")
```

```python
separate_audio_chunk('../assets/lecun_world_model_discussion.mp3')
```

```python
import os
from pydub import AudioSegment

def slice_audio(audio_path, minutes=20):
    """
    Slice audio to keep only the first N minutes.
    
    Args:
        audio_path: Path to the input audio file
        minutes: Number of minutes to keep (default 20)
    
    Exports sliced audio as audio_name_sliced.mp3
    """
    if not os.path.exists(audio_path):
        raise FileNotFoundError(f"Audio file not found: {audio_path}")
    
    if minutes <= 0:
        raise ValueError("minutes must be positive")
    
    # Get base name and directory
    base_name = os.path.basename(audio_path)
    name, _ = os.path.splitext(base_name)
    output_dir = os.path.dirname(audio_path) or "."
    
    # Load audio
    audio = AudioSegment.from_file(audio_path)
    target_duration_ms = minutes * 60 * 1000
    
    # Slice to the target duration
    sliced_audio = audio[:target_duration_ms]
    
    # Export
    output_path = os.path.join(output_dir, f"{name}_sliced_{minutes}_minutes.mp3")
    sliced_audio.export(output_path, format="mp3")
    print(f"Sliced audio ({minutes} min): {output_path}")
```

```python
slice_audio('../assets/lecun_world_model_discussion.mp3', minutes= 5)
```
